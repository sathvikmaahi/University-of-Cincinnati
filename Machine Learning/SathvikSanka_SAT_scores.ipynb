```python
#import statements
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
#read in the SAT.CSV file and store it in data
data = pd.read_csv(r'/content/SAT.csv')
```

```python
#copy the SAT column in data to x and the GPA column to y
x= data['SAT']
y= data['GPA']
```

```python
#visualize it. the complete code is provided
plt.scatter(x,y)
plt.xlabel('SAT')
plt.ylabel('GPA')
```

```python
# define your methods to compute the derivative of sum of the square errors
#GPA = theta * SAT
def hypothesis(theta, x):
    return theta * x
```

```python
# Sum of Squared Errors (SSE) cost function
def cost_function(theta, x, y):
    m = len(y)
    return (1 / (2 * m)) * np.sum((hypothesis(theta, x) - y) ** 2)
```

```python
# Derivative of SSE with respect to theta
def derivative_sse(theta, x, y):
    m = len(y)
    return (1 / m) * np.sum((hypothesis(theta, x) - y) * x)
```

```python
# Gradient Descent algorithm
def gradient_descent(x, y, theta, alpha, iterations):
    cost_history = []
    theta_history = []

    for i in range(iterations):
        # Calculate gradient
        gradient = derivative_sse(theta, x, y)

        # Update theta using the gradient
        theta = theta - alpha * gradient

        # Store the cost and theta for each iteration
        cost = cost_function(theta, x, y)
        cost_history.append(cost)
        theta_history.append(theta)

    return theta, cost_history, theta_history
```

```python
# Initialize theta, learning rate (alpha), and number of iterations
theta = 0  # Initial guess for theta
alpha = 0.00000001  # Learning rate
iterations = 1000  # Number of iterations
```

```python
#Use gradient descent to compute the local minimum for theta.
theta_new = 50
theta_old =0
gamma= 1e-6
precision = 1e-7
```

```python
# Run Gradient Descent to find the optimal theta
theta_optimal, cost_history, theta_history = gradient_descent(x, y, theta, alpha, iterations)
```

```python
predicted_GPA = hypothesis(theta_optimal, x)
```

```python
#plot the graph of the original dataset and the predicted GPA
plt.scatter(x, y, label='Original Data')

# Plot the predicted GPA using 'predicted_GPA' instead of 'GPA'
plt.plot(x, predicted_GPA, color='red', label='Predicted GPA')

plt.xlabel('SAT')
plt.ylabel('GPA')
plt.title('Original vs Predicted GPA')
plt.legend()
plt.show()
```
